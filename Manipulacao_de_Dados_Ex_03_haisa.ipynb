#### Tarefa 2. Descubra qual a idade média por ocupação.

*Dica: use a função groupby do Pandas na coluna occupation*

````python
import pandas as pd

users = pd.read_csv('users_dataset.csv', sep="|")

users = users.drop(['user_id', 'zip_code'], axis=1)

users_age = users.groupby(['occupation']) 

users_age.mean().round()

````

#### Tarefa 3. Para cada ocupação descubra a idade mínima e máxima.

````python
min_age = users_age.min().round()
max_age = users_age.max().round()
display(min_age, max_age)
````

#### Tarefa 4. Para cada combinação de ocupação e sexo, calcule a idade média.

Exemplo:

```
occupation     gender
administrator  F         40.638889
               M         37.162791
artist         F         30.307692
               M         32.333333
```

````python
users_sex = users.groupby(['occupation','gender']).mean().round()
display(users_sex)
````

#### Tarefa 5. Para cada ocupação calcule a porcentagem de homens e mulheres.

Exemplo:

```
occupation     gender
administrator  F          45.569620
               M          54.430380
artist         F          46.428571
               M          53.571429
doctor         M         100.000000
educator       F          27.368421
               M          72.631579
```

````python
users_sum = users.groupby(['gender','occupation']).agg({'gender':'count'})
users_occu = users.groupby(['occupation']).count()
users_perc = users_sum.div(users_occu, level='occupation')*100
users_perc.loc[:,'gender']


````
